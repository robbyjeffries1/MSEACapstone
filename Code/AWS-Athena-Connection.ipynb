# Connecting Jupyter to AWS ATHENA

For this connection to work, you will need:
- a AWS account
- a S3 Bucket
- Identity and Access Management (IAM) Access Key ID and Secret Key

```python
# Authentification
import os
os.environ['ATHENA_USER'] = input("Enter you Access Key ID: ")
os.environ['ATHENA_PASSWORD'] = input("Enter your Secret key: ")
s3_staging_dir = input("Enter the path to your S3 Bucket (e.g: s3://path/to/your/bucket) ")
```

```python
## Requirements

'''
!pip install pyathena==1.2.0
'''
```

```python
from pyathena import connect
import pandas as pd


## Athena Connection
athena_conn = connect(aws_access_key_id=os.environ['ATHENA_USER'], ##credentials of aws_access_key_id
                 aws_secret_access_key=os.environ['ATHENA_PASSWORD'], ##credentials of aws_secret_access_key
                 s3_staging_dir=s3_staging_dir, ##where the athena query result saved - checked in S3 ,
                 region_name='us-east-1') ##the region you set for Athena
```

```python
query="SELECT * FROM amazon_reviews_parquet LIMIT 10"
```

```python
df = pd.read_sql(query, athena_conn)
df.head(5)
```

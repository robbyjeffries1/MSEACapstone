## Import Data

```python
import os
import numpy as np
import pandas as pd
```

```python
os.chdir('/Users/robbyjeffries/MSEACapstone/Data')

raw = pd.read_csv('2015_top_1000.csv')

df = pd.DataFrame(np.array(raw), columns=['marketplace','customer_id','review_id','product_id','product_parent','product_title','star_rating','helpful_votes','total_votes','vine','verified_purchase','review_headline','review_body','review_date','year','product_category'])

# df['review_body'].iloc[0]

df = df[['review_body','star_rating']]
```

```python
df.shape
```

## Data Cleaning

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus = []

for i in range(0,df.shape[0]):
    review = re.sub('[^a-zA-Z]', ' ', df['review_body'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Data Transformation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1420)
```

```python
X = cv.fit_transform(corpus).toarray()
y = df.iloc[:, -1].values
```

```python
# Saving Bag of Words dictionary to use later in prediction
import pickle
bow_path = '/Users/robbyjeffries/MSEACapstone/Data/BoW_Sentiment_Model.pk1'
pickle.dump(cv, open(bow_path, "wb"))
```

## Dividing dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Model fitting (Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```
